# Training Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Create instances of all models
models = {
    'Logistic Regression': LogisticRegression(),
    'Naive Bayes': GaussianNB(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
}

# List to store metrics for each model
metrics = {
    'Model': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': [],
}

for name, model in models.items():
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)

    # Calculate metrics
    acc = accuracy_score(y_test, ypred)
    precision = precision_score(y_test, ypred, average='weighted')
    recall = recall_score(y_test, ypred, average='weighted')
    f1 = f1_score(y_test, ypred, average='weighted')

    # Store metrics
    metrics['Model'].append(name)
    metrics['Accuracy'].append(acc)
    metrics['Precision'].append(precision)
    metrics['Recall'].append(recall)
    metrics['F1 Score'].append(f1)

    # Print metrics
    print(f"{name}:")
    print(f"  Accuracy: {acc}")
    print(f"  Precision: {precision}")
    print(f"  Recall: {recall}")
    print(f"  F1 Score: {f1}")
    print("==========================================================")

# Display metrics summary
metrics_df = pd.DataFrame(metrics)
print(metrics_df)
```

```python
# Previous code that generated 'models' and 'acc'

# Create a list to store model names
model_names = list(models.keys())

# Create a list to store accuracies for each model
accuracies = metrics['Accuracy']  # Extract accuracies from the metrics dictionary

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.barh(model_names, accuracies, color='skyblue', edgecolor='black')  # Changed 'models' to 'model_names' and 'acc' to 'accuracies'
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.title('Comparison of Model Accuracies')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Predictive System

```python
def recommendation(N,P,k,temperature,humidity,ph,rainfal):
    features = np.array([[N,P,k,temperature,humidity,ph,rainfal]])
    transformed_features = ms.fit_transform(features)
    prediction = rfc.predict(transformed_features)
    print(prediction)
    return prediction[0]
```
